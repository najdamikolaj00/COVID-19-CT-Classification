```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model_name = 'EnhancedCNN'
batch_size = 10
epochs = 50
```

```python
data_train = pd.read_csv(f'results/{model_name}_epoch{epochs}_batch{batch_size}_train_results.txt', delimiter=', ')
data_val = pd.read_csv(f'results/{model_name}_epoch{epochs}_batch{batch_size}_val_results.txt', delimiter=', ')
data_test = pd.read_csv(f'results/{model_name}_epoch{epochs}_batch{batch_size}_test_results.txt', delimiter=', ')
```

```python
data_train.columns
```

```python
epochs = data_train['Epoch']
avg_train_loss = data_train['Average Train Loss']
avg_val_loss = data_val['Average Val Loss']

# Plot average train and validation loss over epochs
plt.plot(epochs, avg_train_loss, label='Average Train Loss')
plt.plot(epochs, avg_val_loss, label='Average Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title(f'{model_name} Average Train and Validation Loss')
plt.legend()
plt.show()
```

```python
data_val.columns
```

```python
print(f'{model_name} validation avg. accuracy: ', data_val['Val Acc'].mean())
print(f'{model_name} validation avg. f1 score: ', data_val['F1 Score'].mean())
print(f'{model_name} validation avg. auc: ', data_val['AUC'].mean())
```

```python
print(f'{model_name} test accuracy: ', data_test['Accuracy'])
print(f'{model_name} test f1 score: ', data_test['F1 Score'])
print(f'{model_name} test auc: ', data_test['AUC'])
```
