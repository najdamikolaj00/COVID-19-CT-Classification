```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model_name = 'DenseNet169'
k_fold = 3
batch_size = 10
epochs = 50
```

```python
data_train = pd.read_csv(f'results/{model_name}_k{k_fold}_epoch{epochs}_batch{batch_size}_train_results.txt', delimiter=', ')
data_val = pd.read_csv(f'results/{model_name}_k{k_fold}_epoch{epochs}_batch{batch_size}_val_results.txt', delimiter=', ')
```

```python
data_train.columns
```

```python
num_folds = k_fold
for fold in range(num_folds):
    fold_data = data_train[data_train['Fold'] == fold + 1]
    fold_epochs = fold_data['Epoch']
    fold_train_loss = fold_data['Average Train Loss']
    plt.plot(fold_epochs, fold_train_loss, label=f'Fold {fold + 1} Average Train Loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title(f'{model_name} Average Train Loss for Each Fold')
plt.legend()
plt.show()
```

```python
num_folds = k_fold
for fold in range(num_folds):
    fold_data = data_val[data_val['Fold'] == fold + 1]
    fold_epochs = fold_data['Epoch']
    fold_train_loss = fold_data['Average Val Loss']
    plt.plot(fold_epochs, fold_train_loss, label=f'Fold {fold + 1} Average Validation Loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title(f'{model_name} Average Validation Loss for Each Fold')
plt.legend()
plt.show()
```
